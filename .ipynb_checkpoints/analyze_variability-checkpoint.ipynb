```python
import pandas as pd

# Charger le fichier CSV généré par generate_and_run_all.py
df = pd.read_csv("variability_results.csv")

# Afficher les premières lignes
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogramme du pourcentage de réussite
sns.histplot(df['percent'], bins=20)
plt.title("Distribution du pourcentage de réussite")
plt.xlabel("% réussite")
plt.show()

# Scatter plot pour voir l'effet de n et range_max
sns.scatterplot(data=df, x='n', y='percent', hue='range_max', style='seed')
plt.title("Impact de n et range_max sur le résultat")
plt.show()
```
